```python
# data imports
data = pd.read_csv('./output/YEARLY_PIA_DF_2024.csv')
print(data.count)
print(data.columns)
```

```python
grouped_org = data.groupby(['msr_prd_id', 'org_ip_id', 'end_ivstr_nm', 'scr_code', 'SECTOR', 'ISIC4_CLID', 'BOP_item', 'scr_currencyID', 'issuer_countryID', 'scr_termID'])['orig_ccy_cost_val'].sum().reset_index()
grouped_org.sort_values(by=[ 'org_ip_id', 'end_ivstr_nm', 'scr_code', 'SECTOR', 'ISIC4_CLID', 'BOP_item', 'scr_currencyID', 'issuer_countryID',  'scr_termID', 'msr_prd_id'], inplace=True)

#fix dimension
grouped_org['SECTOR'] = grouped_org['SECTOR'].astype('int').astype('str')
grouped_org['ISIC4_CLID'] = grouped_org['ISIC4_CLID'].astype('int').astype('str')
grouped_org['BOP_item'] = grouped_org['BOP_item'].astype('int').astype('str').str.zfill(10)
grouped_org['scr_currencyID'] = grouped_org['scr_currencyID'].astype('int').astype('str')
grouped_org['issuer_countryID'] = grouped_org['issuer_countryID'].astype('int').astype('str')
grouped_org['scr_termID'] = grouped_org['scr_termID'].astype('int').astype('str')

print(grouped_org.head())
print(grouped_org.info())
```

```python
provider_check = grouped_org.groupby(['msr_prd_id', 'org_ip_id'])['orig_ccy_cost_val'].sum().reset_index()

pivot_provider = provider_check.pivot(index = 'org_ip_id', columns = 'msr_prd_id', values = 'orig_ccy_cost_val' ).reset_index()
print(pivot_provider.head(10))
print(pivot_provider.info())
```

```python
provider_fix = pivot_provider[pivot_provider.iloc[:, 1:].isna().any(axis=1)]
provider_fix.head(10)
```

```python
pivot_grouped_org = grouped_org.pivot(index = ['org_ip_id', 'end_ivstr_nm', 'scr_code', 'SECTOR', 'ISIC4_CLID', 'BOP_item', 'scr_currencyID', 'issuer_countryID', 'scr_termID'], columns = 'msr_prd_id', values = 'orig_ccy_cost_val' ).reset_index()
pivot_grouped_org.head(10)
print(pivot_grouped_org.shape)
```

```python
provider_fix = pivot_grouped_org.iloc[:,:9].merge(pivot_provider, left_on='org_ip_id', right_on = 'org_ip_id')
print(provider_fix.head(10))
print(provider_fix.isna())
print(provider_fix.shape)
```

```python
for col_idx in range(10, len(pivot_grouped_org.columns)):
    for row_idx in range(len(pivot_grouped_org.index)):
        if pd.isna(provider_fix.iloc[row_idx, col_idx]):
            pivot_grouped_org.iloc[row_idx, col_idx] = pivot_grouped_org.iloc[row_idx, col_idx-1]
            
pivot_grouped_org.iloc[:, 9:] = pivot_grouped_org.iloc[:, 9:].fillna(0)
pivot_grouped_org.head(10)
print(pivot_grouped_org.info())
```

```python
### mkt val
grouped_org_mkt = data.groupby(['msr_prd_id', 'org_ip_id', 'end_ivstr_nm', 'scr_code', 'SECTOR', 'ISIC4_CLID', 'BOP_item', 'scr_currencyID', 'issuer_countryID', 'scr_termID'])['orig_ccy_mkt_val'].sum().reset_index()
grouped_org_mkt.sort_values(by=[ 'org_ip_id', 'end_ivstr_nm', 'scr_code', 'SECTOR', 'ISIC4_CLID', 'BOP_item', 'scr_currencyID', 'issuer_countryID',  'scr_termID', 'msr_prd_id'], inplace=True)

grouped_org_mkt['SECTOR'] = grouped_org_mkt['SECTOR'].astype('int').astype('str')
grouped_org_mkt['ISIC4_CLID'] = grouped_org_mkt['ISIC4_CLID'].astype('int').astype('str')
grouped_org_mkt['BOP_item'] = grouped_org_mkt['BOP_item'].astype('int').astype('str').str.zfill(10)
grouped_org_mkt['scr_currencyID'] = grouped_org_mkt['scr_currencyID'].astype('int').astype('str')
grouped_org_mkt['issuer_countryID'] = grouped_org_mkt['issuer_countryID'].astype('int').astype('str')
grouped_org_mkt['scr_termID'] = grouped_org_mkt['scr_termID'].astype('int').astype('str')

provider_check_mkt = grouped_org_mkt.groupby(['msr_prd_id', 'org_ip_id'])['orig_ccy_mkt_val'].sum().reset_index()
pivot_provider_mkt = provider_check_mkt.pivot(index = 'org_ip_id', columns = 'msr_prd_id', values = 'orig_ccy_mkt_val' ).reset_index()
provider_fix_mkt = pivot_provider_mkt[pivot_provider.iloc[:, 1:].isna().any(axis=1)]
pivot_grouped_org_mkt = grouped_org_mkt.pivot(index = ['org_ip_id', 'end_ivstr_nm', 'scr_code', 'SECTOR', 'ISIC4_CLID', 'BOP_item', 'scr_currencyID', 'issuer_countryID', 'scr_termID'], columns = 'msr_prd_id', values = 'orig_ccy_mkt_val' ).reset_index()
provider_fix_mkt = pivot_grouped_org_mkt.iloc[:,:9].merge(pivot_provider, left_on='org_ip_id', right_on = 'org_ip_id')

for col_idx in range(10, len(pivot_grouped_org_mkt.columns)):
    for row_idx in range(len(pivot_grouped_org_mkt.index)):
        if pd.isna(provider_fix_mkt.iloc[row_idx, col_idx]):
            pivot_grouped_org_mkt.iloc[row_idx, col_idx] = pivot_grouped_org_mkt.iloc[row_idx, col_idx-1]
            
pivot_grouped_org_mkt.iloc[:, 9:] = pivot_grouped_org_mkt.iloc[:, 9:].fillna(0)

columns_to_unpivot_mkt = grouped_org_mkt['msr_prd_id'].unique().tolist()
df_unpivoted_mkt = pd.melt(pivot_grouped_org_mkt, 
                       id_vars=['org_ip_id', 'end_ivstr_nm', 'scr_code', 'SECTOR', 'ISIC4_CLID', 'BOP_item', 'scr_currencyID', 'issuer_countryID', 'scr_termID'],
                       value_vars = columns_to_unpivot_mkt,
                       value_name ='orig_ccy_mkt_val')
```

```python
columns_to_unpivot = grouped_org['msr_prd_id'].unique().tolist()
df_unpivoted = pd.melt(pivot_grouped_org, 
                       id_vars=['org_ip_id', 'end_ivstr_nm', 'scr_code', 'SECTOR', 'ISIC4_CLID', 'BOP_item', 'scr_currencyID', 'issuer_countryID', 'scr_termID'],
                       value_vars = columns_to_unpivot,
                       value_name ='orig_ccy_cost_val')
```

```python
df_unpivoted_fin = df_unpivoted
df_unpivoted_fin['orig_ccy_mkt_val'] = df_unpivoted_mkt['orig_ccy_mkt_val']
df_unpivoted_fin.head(3)
# df_unpivoted_fin.sample(3)
```

```python
df_unpivoted_fin.sort_values(by=[ 'org_ip_id', 'end_ivstr_nm', 'scr_code', 'SECTOR', 'ISIC4_CLID', 'BOP_item', 'scr_currencyID',  'scr_termID', 'msr_prd_id'], inplace=True)
df_unpivoted_fin['total_change'] = df_unpivoted_fin.groupby([ 'org_ip_id', 'end_ivstr_nm', 'scr_code', 'SECTOR', 'ISIC4_CLID', 'BOP_item', 'scr_currencyID',  'scr_termID'])['orig_ccy_cost_val'].diff()
df_unpivoted_fin['grand_total_change'] = df_unpivoted_fin.groupby(['msr_prd_id', 'SECTOR', 'ISIC4_CLID', 'BOP_item', 'scr_currencyID', 'issuer_countryID', 'scr_termID'])['total_change'].transform('sum')
df_unpivoted_fin['grand_increase'] = df_unpivoted_fin[df_unpivoted_fin['total_change'] >= 0].groupby(['msr_prd_id', 'SECTOR', 'ISIC4_CLID', 'BOP_item', 'scr_currencyID', 'issuer_countryID', 'scr_termID'])['total_change'].transform('sum')
df_unpivoted_fin['grand_decrease'] = df_unpivoted_fin[df_unpivoted_fin['total_change'] < 0].groupby(['msr_prd_id', 'SECTOR', 'ISIC4_CLID', 'BOP_item', 'scr_currencyID', 'issuer_countryID', 'scr_termID'])['total_change'].transform('sum')
df_unpivoted_fin['transaction_change'] = np.where((df_unpivoted_fin['total_change'] * df_unpivoted_fin['grand_total_change'] >= 0) & (df_unpivoted['grand_total_change'] >= 0),
    df_unpivoted_fin['total_change'] * df_unpivoted_fin['grand_total_change'] / df_unpivoted_fin['grand_increase'],
    np.where((df_unpivoted_fin['total_change'] * df_unpivoted_fin['grand_total_change'] >= 0) & (df_unpivoted_fin['grand_total_change'] < 0),
        df_unpivoted_fin['total_change'] * df_unpivoted_fin['grand_total_change'] / df_unpivoted_fin['grand_decrease'], 0))
df_unpivoted_fin['other_change'] = df_unpivoted_fin['total_change'] - df_unpivoted_fin['transaction_change']

df_unpivoted_fin.head(20)
print(df_unpivoted_fin.info())
print(df_unpivoted_fin.shape)
```

```python
imputed_provider = df_unpivoted_fin.groupby(['org_ip_id','msr_prd_id'])['orig_ccy_cost_val'].sum().reset_index()
pivot_imputed_provider = imputed_provider.pivot(index = ['org_ip_id'], columns = 'msr_prd_id', values = 'orig_ccy_cost_val' ).reset_index()
filtered_df = pivot_imputed_provider[pivot_imputed_provider.isna().any(axis=1)]

filtered_df
```
